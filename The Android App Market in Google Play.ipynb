```python
import pandas as pd
apps_duplicates=pd.read_csv("apps.csv")
apps_duplicates
```

```python
apps=apps_duplicates.drop_duplicates()
print("total no. of apps in the dataset:",len(apps))
```

```python
n=5
apps.sample(n)
```

```python
print("Column headers from list(apps.columns.values):",
      list(apps.columns.values))
```

```python
char_rem=('+',',','$','M')
col_clean=['Installs','Price']
for i in col_clean:
    for j in char_rem:
        apps[i]=apps[i].astype(str).str.replace(j, '');
    apps[i]=pd.to_numeric(apps[i])
apps
```

```python
import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.graph_objs as go
```

```python
num_cat=len(apps["Category"].unique())
print('no. of categories=',num_cat)
```

```python
num_appsincat=apps['Category'].value_counts().sort_values(ascending=False)
data=[go.Bar(x=num_appsincat.index,y=num_appsincat.values)]
plotly.offline.iplot(data)
```

```python
avg_app_rating=apps["Rating"].mean()
print("avg app rating=",avg_app_rating)
data=[go.Histogram(x=apps['Rating'])]
layout={'shapes':[{'type' :'line',
                   'x0':avg_app_rating,
                   'y0':0,'x1':avg_app_rating,
                   'y1':1000,
                  'line':{'dash':'dashdot'}
                  }]}
plotly.offline.iplot({'data':data,'layout':layout})
```

```python
%matplotlib inline
import seaborn as sns
sns.set_style('darkgrid')
import warnings
warnings.filterwarnings('ignore')
la_cat=apps.groupby(apps['Category']).filter(lambda x:len(x)>=250)
plt1=sns.jointplot(x=la_cat['Size'],y=la_cat['Rating'])
paid_apps=apps[apps['Type']=='Paid']
plt2=sns.jointplot(x=paid_apps['Price'],y=paid_apps['Rating'])
```

```python
import matplotlib.pyplot as plt
fig, ax=plt.subplots()
pop_app_cat= apps[apps.Category.isin(['GAME','FAMILY','PHOTOGRAPHY','MEDICAL', 'TOOLS', 'FINANCE','LIFESTYLE','BUSINESS'])]
ax= sns.stripplot(x=pop_app_cat['Price'],y=pop_app_cat['Category'])
ax.set_title('App pricing trend across categories')
apps_above_200=pop_app_cat[['Category','App','Price']][pop_app_cat['Price']>200]
apps_above_200
```

```python
apps_under_100=pop_app_cat[pop_app_cat['Price']<100]
fig, ax=plt.subplots()
fig.set_size_inches(15,8)
ax=sns.stripplot(x=apps_under_100['Price'],y=apps_under_100['Category'],data=apps_under_100)
ax.set_title("app pricing trend across categories after filtering for junk apps")
```

```python
trace0=go.Box(y=apps[apps['Type']=='Paid']['Installs'],name='Paid')
trace1=go.Box(y=apps[apps['Type']=='Free']['Installs'],name='Free')
layout=go.Layout(title='No of downloads of paid apps vs free apps',yaxis=dict(type='log',autorange=True))
data=[trace0,trace1]
plotly.offline.iplot({'data':data,'layout':layout})
```

```python
reviews_df=pd.read_csv("user_reviews.csv")
merged_df=pd.merge(apps,reviews_df,on='App',how='inner')
merged_df=merged_df.dropna(subset={'Sentiment','Translated_Review'})
sns.set_style('ticks')
fig, ax=plt.subplots()
fig.set_size_inches(11,8)
ax=sns.boxplot(x='Type',y='Sentiment_Polarity',data=merged_df)
ax.set_title('Sentiment Polarity Distribution')
```
